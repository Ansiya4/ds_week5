```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr = np.random.randint(50,100,10)
arr
```

```python
x= np.arange(len(arr))
```

```python
plt.bar(x,arr);
```

```python
plt.hist(arr, bins=10)  # Plot the histogram of the values in 'arr'
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Numbers')
plt.show()
```
